```python
import pandas as pd
import gzip
```

```python
file_path_1 = 'E_coli_K12_MG1655.ptt.gz'
file_path_2 = 'B_subtilis_168.ptt.gz'
file_path_3 = 'Halobacterium_NRC1.ptt.gz'
file_path_4 = 'Synechocystis_PCC6803_uid159873.ptt.gz'
```

```python
columns = ["Location", "Strand", "Length", "PID", "Gene", "Synonym", "Code", "COG", "Product"]
```

```python
def zip_ptt(file_path, columns):
    with gzip.open(file_path, 'rt') as f:
        df = pd.read_csv(f, sep='\t', skiprows=3, names=columns)
    return df
```

```python
ecoli_df = zip_ptt(file_path_1, columns)
bsubtilis_df = zip_ptt(file_path_2, columns)
halobacterium_df = zip_ptt(file_path_3, columns)
synechocystis_df = zip_ptt(file_path_4, columns)
```

```python
split_4 = synechocystis_df['Location'].str.extract(r'(\d+)\.\.(\d+)')
synechocystis_df['Start'] = pd.to_numeric(split_4[0])
synechocystis_df['End'] = pd.to_numeric(split_4[1])
```

```python
def operons(df):
    operons_dict = {}
    current_operon_genes = [df.iloc[0].to_dict()]  # Start with the first gene
    operon_count = 1

    for _, row in df.iloc[1:].iterrows():
        last_gene = current_operon_genes[-1]
        # Calculate the intergenic distance
        intergenic_distance = row['Start'] - last_gene['End']

        # Check if the gene is on the same strand and close to the last gene in the current operon
        if row['Strand'] == last_gene['Strand'] and intergenic_distance <= 50:
            current_operon_genes.append(row.to_dict())
        else:
            # Add to operons_dict if current_operon_genes has 2 or more genes
            if len(current_operon_genes) >= 2:
                operons_dict[f"Operon {operon_count}"] = current_operon_genes
                operon_count += 1  # Increment operon count for the next operon
            current_operon_genes = [row.to_dict()]  # Start a new operon

    # Check and add the last operon if it has 2 or more genes
    if len(current_operon_genes) >= 2:
        operons_dict[f"Operon {operon_count}"] = current_operon_genes

    return operons_dict
```

```python
ecoli_operon = operons(ecoli_df)
ecoli_operon
```
